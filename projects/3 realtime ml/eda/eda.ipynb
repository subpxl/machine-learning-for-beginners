```python
import pandas as pd 

df = pd.read_csv("../data/creditcard.csv")

df.describe()
```

```python
df.isnull()
df.isna()

```

```python
y = df["Class"]
x = df.drop(df.columns[-1], axis=1)
x.columns = [f"f{i}" for i in range(x.shape[1])]


from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=43)


import numpy as np
y_train = np.array(y_train).ravel()
y_test = np.array(y_test).ravel()
```

```python
import numpy as np
from sklearn.svm import SVC

import xgboost as xgb
from sklearn.metrics import accuracy_score

model =xgb.XGBClassifier(
    n_estimators=100,
    max_depth =3,
    learning_rate=0.1,
    objective="binary:logistic"  ,
    eval_metric='logloss',
)

model.fit(x_train,y_train)
```

```python
ypred =    model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,roc_auc_score,confusion_matrix
print(accuracy_score(y_test,ypred))

print(classification_report(y_test,ypred   ))
print(roc_auc_score(y_test,ypred   ))
print(confusion_matrix(y_test,ypred))
```

```python
import joblib
joblib.dump(model,"../models/fraud_model.pkl")
model.save_model("../models/fraud_model.json")
```

```python
from onnxmltools import convert_xgboost
from onnxmltools.convert.common.data_types import FloatTensorType
initial_type = [
    ('float_input', FloatTensorType([None, 30]))
]
onnx_model = convert_xgboost(
    model,
    initial_types=initial_type
)

with open("../models/fraud_model.onnx", "wb") as f:
    f.write(onnx_model.SerializeToString())
```

```python
import onnxruntime as rt
import numpy as np

sess = rt.InferenceSession("../models/fraud_model.onnx")

input_name = sess.get_inputs()[0].name

sample = x_test.iloc[0].values.astype(np.float32)

sample = sample.reshape(1, -1)

pred = sess.run(
    None,
    {input_name: sample}    
)

print(pred)
```
